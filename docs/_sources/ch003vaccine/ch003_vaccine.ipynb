```python
import pandas as pd
from IPython.display import Markdown as md
pd.set_option('display.max_rows', 200, 'display.max_colwidth', None)
df=pd.read_csv("../suppfiles/lstcurrent.csv")

df['Issue Date0']= pd.to_datetime(df['Issue Date0'])
lst2=df.drop(['Unnamed: 0','Issue Date'], axis=1).set_index(['Summary','Topic','FDA Organization']).filter(regex='Vaccine|Vaccines',axis=0).filter(regex='^((?!COVID).)*$',axis=0).reset_index()

md("There are "+str(len(lst2))+" documents with topic or summary including 'Vaccine'. Below is a list of these topics.")
```
