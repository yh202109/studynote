```python
import pandas as pd
from IPython.display import Markdown as md
pd.set_option('display.max_rows', 200, 'display.max_colwidth', None)
df=pd.read_csv("../suppfiles/lstcurrent.csv")

lst2=df.drop(['Unnamed: 0'], axis=1).set_index(['Topic','FDA Organization']).filter(like='ICH',axis=0).filter(regex='\(CDER\)',axis=0).reset_index()

md("There are "+str(len(lst2))+" documents with topic including 'ICH'. Below is a list of these topics.")
```

```python
lst2.set_index(['Topic','Summary']).sort_values(['Topic','Summary']).reset_index(['Summary']).drop(['Open for Comment','Docket Number','Issue Date0'], axis=1)
```
